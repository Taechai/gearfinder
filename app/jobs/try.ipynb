```python
from celery import Celery
import time

app = Celery('tasks', broker='pyamqp://localhost:5672')

@app.task
async def task_t(job_id:int):
    # getting the task id
    print("In Progress")
    time.sleep(3)
    print("Job Done:", job_id)
```

```python


async def listen_for_new_jobs():
     pending_jobs = [1, 2, 3, 4, 5, 6]
     for job_id in pending_jobs:
        # Enqueue the job for processing
        task_t.delay(job_id)
        time.sleep(1)
        print("First Job")

if __name__ == "__main__":
    await listen_for_new_jobs()
```
